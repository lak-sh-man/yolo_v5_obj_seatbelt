# Data Cleaning

```python
import os
from glob import glob
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

```python
# load all xml files and store in a list
xml_list = glob('./data_images/*.xml')
# replace \\ with /
xml_list = list(map(lambda x:  x.replace('\\', '/'), xml_list))
```

```python
print(xml_list)
```

```python
# step-2 read xml file
# from each xml file, we need to extract 
# filename, size(width, height), object(name, xmin, xmax, ymin, ymax)
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()
    
    # extract file name 
    image_name = root.find('filename').text
    # width and height of the image 
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser = []
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])
        
    return parser
```

```python
parser_all = list(map(extract_text, xml_list))
```

```python
data = reduce(lambda x, y : x+y, parser_all)
```

```python
len(data)
```

```python
df = pd.DataFrame(data, columns = ['filename', 'width', 'height', 'name', 'xmin', 'xmax', 'ymin', 'ymax' ])
```

```python
# type conversion
cols = ['width', 'height', 'xmin', 'xmax', 'ymin', 'ymax' ]
df[cols] = df[cols].astype(int)
df.info()
```

```python
# center x, center y
df['center_x'] = ((df['xmax']+df['xmin'])/2)/df['width']
df['center_y'] = ((df['ymax']+df['ymin'])/2)/df['height']
# w
df['w'] = (df['xmax']-df['xmin'])/df['width']
# h
df['h'] = (df['ymax']-df['ymin'])/df['height']
```

# Splitting data into Train & Test

```python
images = df['filename'].unique()
```

```python
len(images)
```

```python
# 80% train
# 20% test
img_df = pd.DataFrame(images, columns = ['filename'])
img_train = tuple(img_df.sample(frac = 0.8)['filename']) # shuffle and pick 80% of images
```

```python
print(img_train)
```

```python
img_test = tuple(img_df.query(f'filename not in {img_train}')['filename']) # shuffle and pick 20% of images
```

```python
print(img_test)
```

```python
print(len(img_train))
print(len(img_test))
```

```python
train_df = df.query(f'filename in {img_train}')
test_df = df.query(f'filename in {img_test}')
```

```python
train_df.head()
```

```python
test_df.head()
```

# Label Encoding - Assigning numbers to object names

```python
def label_encoding(x):
    labels = {'With Seatbelt':0, 'Without Seatbelt':1}
    return labels[x]
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
test_df['id'] = test_df['name'].apply(label_encoding)
```

# Save Images and Labels in text form in their respective folders

```python
import os
from shutil import move
```

```python
train_folder = 'data_images/train'
test_folder = 'data_images/test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
cols = ['filename', 'id', 'center_x', 'center_y', 'w', 'h']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
def save_data(filename, folder_path, group_obj):
    # move images
    src = os.path.join('data_images', filename)
    dst = os.path.join(folder_path, filename)
    move(src, dst) # move img to the destination folder

    # move labels
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0]+'.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename, sep = ' ', index = False, header = False)
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data, args = (train_folder, groupby_obj_train))
```

```python
filename_series_test = pd.Series(groupby_obj_test.groups.keys())
```

```python
filename_series_test.apply(save_data, args = (test_folder, groupby_obj_test))
```
